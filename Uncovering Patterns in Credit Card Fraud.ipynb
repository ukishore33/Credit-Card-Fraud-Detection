### Key Findings from EDA
1. Class Imbalance is Extreme
Fraud is less than 0.2% of the data.

This confirms the need for careful analysis and possibly special techniques in modeling (like oversampling, anomaly detection, etc.).

2. Transaction Amount Analysis
Average amount (Fraud) is higher than normal on average.

However, there's overlap — meaning amount alone can’t determine fraud.

3. Feature Correlation with Fraud
Features like V4, V10, V11, V14, V17 show strong correlation (positive or negative) with fraud cases.

These could be key predictors in a model.

4. Boxplots Reveal Patterns
Boxplots of selected PCA features show clear differences between fraud and normal classes.

Example: V14 often has much lower values in fraud cases.

5. Time Trends
When visualized, fraud seems to cluster around specific time intervals (in seconds from the first transaction).

Could suggest automated attacks or specific fraud time windows.

6. Transaction Count Visualization
Bar plot shows an overwhelming number of normal transactions, visually reinforcing the class imbalance.
